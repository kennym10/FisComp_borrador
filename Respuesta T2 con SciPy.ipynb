```python
from pylab import *
from scipy.integrate import *
```

```python
def ec_prob1(t,x):
    # x es (x,dx) 
    dx = x[1]
    dvx = -0.5*x[1] - sin(x[0]) + exp(-t)*sin(3*t)
    return array([dx,dvx])
```

```python
ci = [0,1]
t_f = 10*pi
solucion_p1 = solve_ivp(ec_prob1,(0,t_f),ci,dense_output=True)
```

```python
t=linspace(0,t_f,500)
plot(t,solucion_p1.sol(t)[0])
show()
```

```python
def ec_prob2(t,x):
    # x es (x,dx,ddx) 
    dx = x[1]
    dvx = x[2]
    dax = -x[2]**2 + 3*x[1]**3 - (cos(x[0]))**2 + exp(-t)*sin(3*t)
    return array([dx,dvx,dax])
```

```python
ci = [1,2,1]
solucion_p2_1 = solve_ivp(ec_prob2,(1,1.976),ci)
solucion_p2_1.success
```

```python
#t=linspace(0,2,100)
plot(solucion_p2_1.t,solucion_p2_1.y[0])
show()
```

```python
def ec_prob3(t,x,omega=1):
    # x es (x,dx) 
    dx = x[1]
    dvx = -0.1*x[1] - 4*sin(x[0]) + 0.5*sin(omega*t)
    return array([dx,dvx])
```

```python
ci = [0,1]
omegas = arange(1,3.1,0.1)
t_f = 200
amplitudes = zeros(len(omegas))
t = linspace(0,t_f,500) 
for i in range(len(omegas)):
    solucion_p1 = solve_ivp(ec_prob3,(0,t_f),ci,args=(omegas[i],),
                            dense_output=True)
    t_finales = linspace(150,t_f,500)
    amplitudes[i] = max(abs(solucion_p1.sol(t_finales)[0]))
    plot(t,solucion_p1.sol(t)[0])
show()
```

```python
plot(omegas,amplitudes)
show()
```

```python
def ec_prob4(t,x):
    # x es (x,y,z) 
    dx = 10*(x[1]-x[0])
    dy = x[0]*(28-x[2]) -x[1]
    dz = x[0]*x[1] - (8/3)*x[2]
    return array([dx,dy,dz])
```

```python
ci = [2,3,1.5]
solucion_p4 = solve_ivp(ec_prob4,(0,30),ci,dense_output=True)
```

```python
t = linspace(0,30,2000)
plot(solucion_p4.sol(t)[0],solucion_p4.sol(t)[2])
show()
```
